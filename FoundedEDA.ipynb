# Importing packages

```python
# Packages / libraries
import os #provides functions for interacting with the operating system
import numpy as np 
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# 1. Loading the Raw Data¶


```python
# Loading the data

movies = pd.read_csv("movies.csv")

#runs the first 5 rows
movies.head()
```

# 2. Line Graph

```python
# The line graph is used to answer the question: What does the relationship between the imdb scores and
# movie_facebook_likes tell about the genres of movies the fans like to watch?

# Example 1 - Simple 1 line graph
# Investigating the correlation of two variables to the moviegoer's film perferences.

ax = sns.lineplot(x='movie_facebook_likes', y='imdb_score', data = movies)

# Notes: error bands show the confidence interval
```

```python
# Example 2 - Adding Categories

# By Color column
ax = sns.lineplot(x='imdb_score', y='movie_facebook_likes', hue = 'color', data = movies)
```

# 3. lmplots

```python
# Example 2 - Relationship between gross revenues and imdb scores to answer the question:
# How does these variable factors describe moviegoers interest in the genres of films?
sns.set(color_codes=True)
ax = sns.lmplot(x="imdb_score", y="gross (in thousands)", height=20, data=movies, palette=dict(Yes="g", No="m"))
```

# 4. Bar plots

```python
# Example 4 - Investigating the Return on Investment (ROI) based on the gross revenues and budgeting to come
# with this question: How will the ROI help to choose which genres of movies to produce?

profits = []
roi_vals = []

for index, row in movies.iterrows():
    profit = row['gross (in thousands)']- row['budget']
    budget = row['budget']
    num = profit - budget
    den = budget
    # convert roi to percentage
    roi = (num / den) * 100
    
    profits.append(profit)
    roi_vals.append(roi)
    
movies['profit'] = profits
movies['roi'] = roi_vals

movies.head()
```
